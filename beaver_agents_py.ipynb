```python
import csv

# ---------------------------
# MOCK FUNCTIONS (replace later if you have project_starter.py)
# ---------------------------

def get_stock_level(product):
    return 200

def search_quote_history(product):
    return "Previous price: 10 per unit"

def create_transaction(product, quantity, price):
    return "Transaction successful"

def get_cash_balance():
    return 50000

def extract_quantity(request):
    words = request.split()
    for word in words:
        if word.isdigit():
            return int(word)
    return 0


# ---------------------------
# AGENTS (Simple Python)
# ---------------------------

class InventoryAgent:
    def run(self, request):
        return f"Stock available: {get_stock_level(request)}"


class QuoteAgent:
    def run(self, request):
        quantity = extract_quantity(request)
        price_per_unit = 10

        if quantity > 100:
            discount = 0.1
        else:
            discount = 0

        total = quantity * price_per_unit * (1 - discount)

        return f"Price: {total} (discount {discount*100}%)"


class SalesAgent:
    def run(self, request, stock):
        quantity = extract_quantity(request)

        if quantity > stock:
            return "Order failed: insufficient stock"
        return "Transaction successful"


class FinanceAgent:
    def run(self):
        return f"Current balance: {get_cash_balance()}"


# ---------------------------
# ORCHESTRATOR
# ---------------------------

def handle_request(request):
    quantity = extract_quantity(request)

    stock = get_stock_level(request)

    inventory = f"Stock available: {stock}"
    quote = QuoteAgent().run(request)
    sale = SalesAgent().run(request, stock)
    finance = FinanceAgent().run()

    return {
        "request": request,
        "inventory": inventory,
        "quote": quote,
        "sale": sale,
        "finance": finance
    }

# ---------------------------
# TEST RUN
# ---------------------------

result = handle_request("A4 paper 500 units")
print(result)
```
